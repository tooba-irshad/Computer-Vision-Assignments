```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('3.JPG')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_float = np.float32(gray)
harris_corners = cv2.cornerHarris(src=gray_float, blockSize=2, ksize=3, k=0.04)
# Step 4: Enhance corner visibility by dilating them
harris_corners = cv2.dilate(harris_corners, None)
# Step 5: Thresholding to mark the corners in the original image
# Mark corners in red
threshold = 0.01 * harris_corners.max()
image[harris_corners > threshold] = [0, 0, 255]

# Step 6: Visualize the detected corners by overlaying them on the original image
# Convert BGR to RGB for visualization in Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Display the original image with corners marked
plt.figure(figsize=(10, 6))
plt.title("Harris Corner Detection")
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
